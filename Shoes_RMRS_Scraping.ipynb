```python
from splinter import Browser
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://ragged-mountain-running-shop.myshopify.com/collections/mens-footwear'
browser.visit(url)
```

```python
# Iterate through all pages
list_of_shoes = []
price_list = []
for x in range(10):
    # HTML object
    html = browser.html
    # Parse HTML with Beautiful Soup
    soup = BeautifulSoup(html, 'html.parser')
    # Retrieve all elements that contain shoe information
    shoes = soup.find_all('div', class_='grid-view-item product-card')

    # Iterate through each shoe
    for shoe in shoes:
        # Use Beautiful Soup's find() method to navigate and retrieve attributes
        #shoe_type = shoe.find('span', class_='visually-hidden').text.strip()
        shoe_type = shoe.find('div', class_='h4 grid-view-item__title product-card__title').text.strip()
        list_of_shoes.append(shoe_type)
        
        price = shoe.find('span', "price-item price-item--regular").text.strip()
        price_list.append(price)

        print(shoe_type)
        print(price)
        print('-----------')
        
    # Click the 'Next' button on each page
    try:
        browser.links.find_by_partial_text('Next page').click()
          
    except:
        print("Scraping Complete")
```

```python
browser.quit()
```
